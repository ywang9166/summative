```python
# Create a list of the 15 out of 42 columns to keep
columns_to_keep = ['interest_rate', 'verified_income', 'debt_to_income',
                   'total_credit_utilized', 'total_credit_limit',
    'public_record_bankrupt', 'loan_purpose', 'term',
    'inquiries_last_12m', 'issue_month', 'annual_income',
    'loan_amount', 'grade', 'emp_length', 'homeownership'
]
```

```python
# filter the dataset
df_filtered = df[columns_to_keep]
```

```python
df_filtered
```

```python
df_filtered = df_filtered.rename(columns={"inquiries_last_12m":"credit checks"})
```

```python
print(df_filtered.columns)
```

```python
df_filtered.describe()
```

```python
# before cleaning, use df.shape to get the number of row, which is the number of
# observations
# df.shape returns a tuple like (rows, columns)
print("before cleaning:", df.shape[0])
```

```python
# Remove all rows with missing (NaN) values from the df_filtered DataFrame.
df_cleaned1 = df_filtered.dropna()
```

```python
print("after cleaning:", df_cleaned1.shape[0])
```

```python
# loop through the 10 numeric columns to filter out outliers using IQR
numeric_columns = ['interest_rate', 'debt_to_income', 'total_credit_utilized',	'total_credit_limit',
                'public_record_bankrupt', 'term', 'credit checks', 'annual_income', 'loan_amount',
                'emp_length']
for column in numeric_columns:
  Q1 = df_cleaned1[column].quantile(0.25)
  Q3 = df_cleaned1[column].quantile(0.75)
  # IQR (interquartile range): the range of the middle 50% of your data
  IQR = Q3 - Q1
  # Use & for element-wise AND, and wrap each condition in parentheses
  # Use IQR to detect outliers， an outlier is defined as any value that falls below
  # Q1 - 1.5IQR or above Q3 + 1.5 IQR
  df_cleaned = df_cleaned1[(df_cleaned1[column] >= Q1 - 1.5*IQR) & (df_cleaned1[column] <= Q3 +1.5*IQR)]
```

```python
df_cleaned
```

```python
print("after cleaning:", df_cleaned.shape[0])
```

```python
# show boxplot
import seaborn as sns
import matplotlib.pyplot as plt

numeric_columns = ['interest_rate', 'debt_to_income', 'total_credit_utilized',	'total_credit_limit',
                'public_record_bankrupt', 'term', 'credit checks', 'annual_income', 'loan_amount',
                'emp_length']
for column in numeric_columns:
  plt.figure(figsize=(10, 6))
  sns.boxplot(data=df_cleaned1, x=column)
  plt.title(f'Boxplot of {column}')
  plt.show()
# Box: interquartile range (IQR: Q1 to Q3)
# Line inside box: the median
# Whiskers: range within 1.5×IQR
# Dots outside: outliers
```
